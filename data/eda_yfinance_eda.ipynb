### Download Sample Stock Data

Let's download historical data for a popular stock, for example, Apple (AAPL).

```python
import pandas as pd
import os

# Define the ticker symbols corresponding to the files in the directory
all_ticker_symbols = [
    'AMZN', 'AAPL', 'GOOGL', 'MSFT', 'UDMY', 'NXE', 'SPY', 'CDR.WA', 'EH',
    'BTC-USD', 'ETH-USD'
]

# Dictionary to store data for each ticker
data = {}

# Load data from local CSV files
for ticker in all_ticker_symbols:
    file_path = f'/content/{ticker}.csv'
    if os.path.exists(file_path):
        print(f"Loading data from {file_path}...")
        # Load the CSV and set the Date column as the index
        df_temp = pd.read_csv(file_path, header=[0, 1], index_col=0, parse_dates=True)
        data[ticker] = df_temp
    else:
        print(f"Warning: File {file_path} not found.")

print("Data loading from files complete.")
```

### Display the first few rows of the data

```python
# Display the first few rows for each downloaded dataset
for ticker, df_ticker in data.items():
    if not df_ticker.empty:
        print(f"\n--- {ticker} Data Head ---")
        display(df_ticker.head())
```

### Check for Missing Timestamps

It's crucial to identify any missing timestamps in the historical data, as gaps can affect analysis. We will iterate through each dataset and check for continuity in the `Date` index.

```python
for ticker, df_ticker in data.items():
    if not df_ticker.empty:
        print(f"\n--- Checking {ticker} for missing timestamps ---")
        # Fix: Use utc=True to handle timezone-aware strings from CSVs
        if not isinstance(df_ticker.index, pd.DatetimeIndex):
            df_ticker.index = pd.to_datetime(df_ticker.index, utc=True)
        elif df_ticker.index.tz is None:
            df_ticker.index = df_ticker.index.tz_localize('UTC')

        # Create a complete date range using the same timezone (UTC)
        full_date_range = pd.date_range(start=df_ticker.index.min(), end=df_ticker.index.max(), freq='D', tz='UTC')

        # Identify missing dates
        missing_dates = full_date_range.difference(df_ticker.index)

        if not missing_dates.empty:
            print(f"  Found {len(missing_dates)} missing dates for {ticker}:")
            print(f"  {missing_dates[:5]}... (showing first 5 if many)")
        else:
            print(f"  No missing dates found for {ticker} within its range.")
    else:
        print(f"\n--- No data to check for {ticker} ---")
```

### Computing Daily Returns

We will calculate the daily percentage change of the 'Close' price for each asset to understand day-to-day volatility.

```python
daily_returns_df = pd.DataFrame()

for ticker, df_ticker in data.items():
    if not df_ticker.empty:
        # Ensure we are using the 'Close' price column
        # yfinance CSVs often have a MultiIndex or specific column naming
        if isinstance(df_ticker.columns, pd.MultiIndex):
            close_prices = df_ticker['Close']
        else:
            close_prices = df_ticker['Close']

        daily_returns_df[ticker] = close_prices.pct_change()

print("Daily returns calculation complete. Here are the first 5 rows:")
display(daily_returns_df.head())
```

### Compare Date Ranges Across Assets

Let's explicitly compare the date ranges for each asset to understand the difference between stock trading days and continuous cryptocurrency trading.

```python
print("\n--- Date Ranges for Each Asset ---")
for ticker, df_ticker in data.items():
    if not df_ticker.empty:
        min_date = df_ticker.index.min()
        max_date = df_ticker.index.max()
        print(f"{ticker}: {min_date.strftime('%Y-%m-%d')} to {max_date.strftime('%Y-%m-%d')}")
    else:
        print(f"{ticker}: No data available.")
```

### Next Step: Check for Missing Values

Now, let's proceed with checking for any missing values within the actual data columns (e.g., Open, High, Low, Close, Volume) for each asset.

```python
print("\n--- Checking for Missing Values in Data Columns ---")
for ticker, df_ticker in data.items():
    if not df_ticker.empty:
        missing_values_count = df_ticker.isnull().sum()
        if missing_values_count.sum() > 0:
            print(f"\n  Missing values for {ticker}:")
            print(missing_values_count[missing_values_count > 0])
        else:
            print(f"\n  No missing values found for {ticker} in data columns.")
    else:
        print(f"\n--- No data to check for missing values for {ticker} ---")
```

### Check for Zero-Volume Rows (Potential Data Artifacts)

Zero-volume rows can sometimes indicate data errors or non-trading days that slipped through, rather than genuine trading inactivity. Let's check for these for each asset.

```python
print("\n--- Checking for Zero-Volume Rows ---")
for ticker, df_ticker in data.items():
    if not df_ticker.empty and 'Volume' in df_ticker.columns:
        zero_volume_rows = df_ticker[df_ticker['Volume'] == 0]
        if not zero_volume_rows.empty:
            print(f"  Found {len(zero_volume_rows)} zero-volume rows for {ticker}:")
            display(zero_volume_rows.head())
        else:
            print(f"  No zero-volume rows found for {ticker}.")
    elif not df_ticker.empty and 'Volume' not in df_ticker.columns:
        print(f"  '{ticker}' DataFrame does not have a 'Volume' column.")
    else:
        print(f"--- No data to check for zero-volume rows for {ticker} ---")
```

### Confirm Data Granularity

It's important to confirm the time granularity of the data (e.g., daily, hourly). `yfinance` typically provides daily data, but let's verify this by checking the frequency of the `DatetimeIndex`.

```python
print("\n--- Confirming Data Granularity ---")
for ticker, df_ticker in data.items():
    if not df_ticker.empty:
        # Ensure the index is a DatetimeIndex
        if not isinstance(df_ticker.index, pd.DatetimeIndex):
            df_ticker.index = pd.to_datetime(df_ticker.index)

        # Infer the frequency
        inferred_freq = pd.infer_freq(df_ticker.index)
        if inferred_freq:
            print(f"  {ticker}: Inferred frequency is {inferred_freq}")
        else:
            print(f"  {ticker}: Could not infer frequency. Checking average time difference.")
            # If frequency cannot be inferred, calculate average difference
            if len(df_ticker.index) > 1:
                time_diffs = df_ticker.index.to_series().diff().dropna()
                avg_diff = time_diffs.mean()
                print(f"    {ticker}: Average time difference between entries: {avg_diff}")
            else:
                print(f"    {ticker}: Only one data point, cannot determine frequency.")
    else:
        print(f"--- No data to check granularity for {ticker} ---")

print("\nBased on the analysis, we can determine if data unification (e.g., aggregating to daily if some are hourly) is necessary. Given that yfinance typically provides daily data, we expect most to be 'D' or similar daily frequencies.")
```

```python
# 1. Robust Unification with Date Normalization
unified_returns_list = []

# Create a normalized reference index (Dates only) from SPY
if 'SPY' in data:
    spy_idx = data['SPY'].index
    market_dates = pd.to_datetime(spy_idx).tz_localize(None).normalize().unique()

    for ticker, df_ticker in data.items():
        # Extract 'Close' column robustly
        if isinstance(df_ticker.columns, pd.MultiIndex):
            close_col = [col for col in df_ticker.columns if col[0] == 'Close'][0]
            prices = df_ticker[close_col]
        else:
            prices = df_ticker['Close']

        # Normalize this asset's index to date-only (removing hour offsets)
        prices_normalized = prices.copy()
        prices_normalized.index = pd.to_datetime(prices_normalized.index).tz_localize(None).normalize()

        # Reindex to SPY dates and calculate returns
        aligned_prices = prices_normalized.reindex(market_dates).ffill()
        returns = aligned_prices.pct_change()
        returns.name = ticker
        unified_returns_list.append(returns)

    unified_df = pd.concat(unified_returns_list, axis=1)
    print(f"Unified dataframe rebuilt with {len(market_dates)} aligned days.")

    # 2. Re-calculate and Plot Volatility
    vol_data = unified_df.std() * np.sqrt(252) * 100
    vol_data = vol_data.sort_values(ascending=False)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=vol_data.index, y=vol_data.values, palette='magma', hue=vol_data.index, legend=False)
    plt.title('Final Corrected Annualized Volatility (All 11 Assets)', fontsize=15)
    plt.ylabel('Annualized Volatility (%)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    for i, val in enumerate(vol_data.values):
        if not np.isnan(val):
            plt.text(i, val + 0.5, f'{val:.1f}%', ha='center', fontweight='bold')
    plt.show()
else:
    print("Error: SPY not found for reference.")
```

### Drawdown Analysis: Visualizing Portfolio Risk
Drawdown measures the peak-to-trough decline during a specific period. It is a key metric for understanding the risk and potential 'pain' an investor might experience holding volatile assets like Cryptocurrencies or speculative stocks.

```python
import matplotlib.pyplot as plt

# Identify assets with highest volatility from earlier analysis
risky_assets = ['BTC-USD', 'ETH-USD', 'NXE']

# Calculate cumulative returns
cum_returns = (1 + unified_df[risky_assets].fillna(0)).cumprod()

# Calculate running maximums
running_max = cum_returns.cummax()

# Calculate Drawdown as a percentage
drawdowns = (cum_returns - running_max) / running_max * 100

# Plotting the Drawdown Timeline
plt.figure(figsize=(14, 7))
for ticker in risky_assets:
    plt.plot(drawdowns.index, drawdowns[ticker], label=f'{ticker} Drawdown')
    plt.fill_between(drawdowns.index, drawdowns[ticker], 0, alpha=0.15)

plt.title('Drawdown Timeline: Riskiest Assets (BTC, ETH, NXE)', fontsize=16)
plt.ylabel('Percentage Drop from Peak (%)')
plt.xlabel('Date')
plt.legend(loc='lower left')
plt.axhline(0, color='black', lw=1)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# Calculate Max Drawdown for context
max_dd = drawdowns.min()
print("Maximum Drawdowns encountered in this period:")
for ticker, value in max_dd.items():
    print(f"{ticker}: {value:.2f}%")
```

### Data Foundations: Coverage and Completeness
To ensure the robustness of our analysis, we validate the data by checking the date coverage, granularity, and the presence of missing values (NaNs) or artifacts (zero volume) for all 11 assets.

```python
import pandas as pd
import numpy as np

coverage_stats = []

for ticker in all_ticker_symbols:
    if ticker in data:
        df_ticker = data[ticker]
        idx = df_ticker.index

        # Extracting metrics
        start_date = idx.min().strftime('%Y-%m-%d')
        end_date = idx.max().strftime('%Y-%m-%d')
        total_rows = len(df_ticker)

        # Infer frequency
        freq = pd.infer_freq(idx) if len(idx) > 1 else 'N/A'
        if not freq:
            freq = 'Daily (Inferred)'

        # Calculate missing values % in 'Close' column
        if isinstance(df_ticker.columns, pd.MultiIndex):
            close_col = [col for col in df_ticker.columns if col[0] == 'Close'][0]
            nulls = df_ticker[close_col].isnull().sum()
        else:
            nulls = df_ticker['Close'].isnull().sum()

        # Ensure nulls is a scalar
        if hasattr(nulls, 'iloc'): nulls = nulls.iloc[0]
        missing_pct = (nulls / total_rows) * 100

        # Calculate zero-volume rows
        if 'Volume' in df_ticker.columns:
            # Handle potential multi-index for Volume as well
            vol_data = df_ticker['Volume']
            zeros = (vol_data == 0).sum()
            if hasattr(zeros, 'iloc'): zeros = zeros.iloc[0]
            zero_pct = (zeros / total_rows) * 100
        else:
            zero_pct = np.nan

        coverage_stats.append({
            'Asset': ticker,
            'Start Date': start_date,
            'End Date': end_date,
            'Rows': total_rows,
            'Granularity': freq,
            'Missing Close %': f'{float(missing_pct):.2f}%',
            'Zero Volume %': f'{float(zero_pct):.2f}%' if not pd.isna(zero_pct) else 'N/A'
        })

coverage_df = pd.DataFrame(coverage_stats)

print("Data Coverage and Completeness Summary:")
display(coverage_df)
```

### Volatility Comparison (Annualized)
To compare risk across different asset classes (Stocks vs. Crypto), we calculate the **Annualized Volatility**.
* For stocks, we typically use 252 trading days.
* Since we unified our data to stock market days, we will use 252 as the scaling factor for all assets in this specific unified view.

```python
import numpy as np

# Calculate daily standard deviation and annualize it (std * sqrt(252))
volatility = unified_df.std() * np.sqrt(252) * 100
volatility = volatility.sort_values(ascending=False)

# Create the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=volatility.index, y=volatility.values, palette="magma", hue=volatility.index, legend=False)

plt.title('Annualized Volatility Comparison (All 11 Assets)', fontsize=15)
plt.ylabel('Annualized Volatility (%)')
plt.xlabel('Asset Ticker')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add values on top of bars
for i, val in enumerate(volatility.values):
    plt.text(i, val + 0.5, f'{val:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
